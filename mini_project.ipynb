```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import json
import matplotlib.pyplot as plt # we only need pyplot
```

```python
data = pd.read_csv('card_data.csv')
data.head()
```

```python
print(data.dtypes)
```

```python
data = data.drop(['desc', 'pend_desc', 'monster_desc'], axis=1)
data.head()
```

```python
data['card_prices'] = data['card_prices'].str.strip('[]')
# Convert the string in the 'card_prices' column to a dictionary
data['card_prices'] = data['card_prices'].apply(lambda x: json.loads(x.replace("'", '"')))

# Normalize the 'card_prices' column to create separate columns for each key in the dictionary
data = pd.concat([data, pd.json_normalize(data['card_prices'])], axis=1)

# Now the DataFrame 'df' contains 5 new columns with the prices from the 'card_prices' column
data.head()
```

```python
unique_types = data['type'].unique()
print(unique_types)
```

```python
data = data[~data['type'].isin(['Spell Card', 'Trap Card', 'Token'])]
data.head()
```

```python
# Define the bins for the level ranges
bins = [1.0, 4.0, 6.0, float('inf')]

# Define the labels for the level ranges
labels = ['0', '1', '2']

# Create a new column 'level_group' to store the bucketed groups
data['tribute'] = pd.cut(data['level'], bins=bins, labels=labels, right=False)

# Now the DataFrame 'df' contains a new column 'level_group' with the cards bucketed into the specified level ranges
data.head()
```
